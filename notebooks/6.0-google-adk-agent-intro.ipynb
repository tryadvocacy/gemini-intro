## Example 1: Basic Agent

Let's create a simple agent that can respond to queries:

```python
from google.adk.agents.llm_agent import Agent

# Create a basic agent
basic_agent = Agent(
    model='gemini-2.0-flash-exp',
    name='basic_assistant',
    description="A helpful AI assistant",
    instruction="You are a helpful assistant. Provide clear and concise answers.",
)

print("Basic agent created successfully!")
```

```python
# Test the basic agent
response = basic_agent.run("What is artificial intelligence?")
print(response)
```

## Example 2: Agent with Custom Tools

Now let's create an agent with custom tools/functions. This demonstrates function calling capabilities:

```python
from datetime import datetime
import pytz

def get_current_time(city: str) -> dict:
    """
    Returns the current time in a specified city.
    
    Args:
        city: The name of the city (e.g., 'New York', 'London', 'Tokyo')
    
    Returns:
        A dictionary with status, city, and current time
    """
    # Mapping of cities to timezones
    city_timezones = {
        'new york': 'America/New_York',
        'london': 'Europe/London',
        'tokyo': 'Asia/Tokyo',
        'paris': 'Europe/Paris',
        'sydney': 'Australia/Sydney',
        'los angeles': 'America/Los_Angeles',
        'chicago': 'America/Chicago',
        'dubai': 'Asia/Dubai',
        'singapore': 'Asia/Singapore',
    }
    
    city_lower = city.lower()
    
    if city_lower in city_timezones:
        tz = pytz.timezone(city_timezones[city_lower])
        current_time = datetime.now(tz).strftime('%I:%M %p %Z')
        return {
            "status": "success",
            "city": city,
            "time": current_time
        }
    else:
        return {
            "status": "error",
            "city": city,
            "message": f"Timezone information not available for {city}"
        }

def calculate(expression: str) -> dict:
    """
    Evaluates a mathematical expression and returns the result.
    
    Args:
        expression: A mathematical expression as a string (e.g., '2 + 2', '10 * 5')
    
    Returns:
        A dictionary with the expression and result
    """
    try:
        # Safe evaluation of mathematical expressions
        result = eval(expression, {"__builtins__": {}}, {})
        return {
            "status": "success",
            "expression": expression,
            "result": result
        }
    except Exception as e:
        return {
            "status": "error",
            "expression": expression,
            "message": str(e)
        }

print("Tool functions defined!")
```

```python
# Create an agent with tools
tool_agent = Agent(
    model='gemini-2.0-flash-exp',
    name='tool_assistant',
    description="An assistant that can tell time and perform calculations",
    instruction="""You are a helpful assistant with access to tools.
    - Use the 'get_current_time' tool when users ask about time in different cities
    - Use the 'calculate' tool when users ask you to perform mathematical calculations
    Provide friendly and informative responses.""",
    tools=[get_current_time, calculate],
)

print("Tool-equipped agent created successfully!")
```

## Testing the Tool Agent

Let's test the agent with queries that require tool usage:

```python
# Test 1: Ask for time in a city
print("Query: What time is it in Tokyo?")
print("-" * 50)
response = tool_agent.run("What time is it in Tokyo?")
print(response)
```

```python
# Test 2: Ask for a calculation
print("Query: What is 1234 multiplied by 5678?")
print("-" * 50)
response = tool_agent.run("What is 1234 multiplied by 5678?")
print(response)
```

```python
# Test 3: Multiple tools in one query
print("Query: What time is it in New York and London? Also calculate 365 * 24")
print("-" * 50)
response = tool_agent.run("What time is it in New York and London? Also calculate 365 * 24")
print(response)
```

## Example 3: Domain-Specific Agent

Let's create a more specialized agent for a specific use case:

```python
def get_weather(city: str) -> dict:
    """
    Returns mock weather information for a city.
    (In production, this would call a real weather API)
    
    Args:
        city: The name of the city
    
    Returns:
        A dictionary with weather information
    """
    # Mock data for demonstration
    import random
    
    conditions = ['Sunny', 'Cloudy', 'Rainy', 'Partly Cloudy']
    temp = random.randint(15, 30)
    
    return {
        "status": "success",
        "city": city,
        "temperature": f"{temp}°C",
        "condition": random.choice(conditions),
        "note": "This is mock data for demonstration purposes"
    }

# Create a travel assistant agent
travel_agent = Agent(
    model='gemini-2.0-flash-exp',
    name='travel_assistant',
    description="A travel planning assistant",
    instruction="""You are a friendly travel planning assistant.
    Help users with:
    - Current time in different cities (use get_current_time)
    - Weather information (use get_weather)
    - Travel tips and recommendations
    Be enthusiastic and helpful!""",
    tools=[get_current_time, get_weather],
)

print("Travel assistant agent created!")
```

```python
# Test the travel agent
query = "I'm planning to visit Paris. What's the weather like there and what time is it now?"
print(f"Query: {query}")
print("-" * 50)
response = travel_agent.run(query)
print(response)
```
